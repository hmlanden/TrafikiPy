```python
#===========DEPENDENCIES=============
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import seaborn as sns
import requests

from datetime import datetime

%matplotlib inline

pd.set_option('display.max_columns', None)
```

```python
#============IMPORT==============
csv_file_path = os.path.join('Resources', 'accidents_2014.csv')
traffic_df = pd.read_csv(csv_file_path)
traffic_df
```

```python
#============DROP BLANK COLUMNS===========

traffic_df.dropna(
    axis=1,
    how='all',
    inplace=True
)

#============DROP BLANK ROWS WITH BLANK VALUES==========

traffic_df['Junction_Control'].replace(
    np.nan, 'None', inplace=True)

traffic_df.replace(
    '', np.nan, inplace=True)

traffic_df.replace(
    'Unknown', np.nan, inplace=True)

traffic_df.dropna(axis=0, inplace=True)

#===========REPLACING ERRANT/MISPELLED VALUES===============

traffic_df['Light_Conditions'].replace(
    'Darkeness: No street lighting',
    'Darkness: No street lighting', 
    inplace=True
)

traffic_df['Pedestrian_Crossing-Physical_Facilities'].replace(
    'non-junction pedestrian crossing',
    'Non-junction Pedestrian Crossing', 
    inplace=True
)

#===========RENAMING COLUMNS===============

traffic_df.rename(columns=
    {'Accident_Index' : 'Accident Index',
     'Longitude' : 'Longitude', 
     'Latitude' : 'Latitude', 
     'Police_Force' : 'Police Force', 
     'Accident_Severity' : 'Accident Severity', 
     'Number_of_Vehicles' : 'Number of Vehicles', 
     'Number_of_Casualties' : 'Number of Casualties', 
     'Date' : 'Date', 
     'Day_of_Week' : 'Day of Week', 
     'Time' : 'Time', 
     'Local_Authority_(District)' : 'Local Authority District', 
     'Local_Authority_(Highway)' : 'Local Authority Highway', 
     '1st_Road_Class' : '1st Road Class', 
     '1st_Road_Number' : '1st Road Number', 
     'Road_Type' : 'Road Type', 
     'Speed_limit' : 'Speed Limit', 
     'Junction_Control' : 'Junction Control', 
     '2nd_Road_Class' : '2nd Road Class', 
     '2nd_Road_Number' : '2nd Road Number', 
     'Pedestrian_Crossing-Human_Control' : 'Pedestrian Crossing Human Control', 
     'Pedestrian_Crossing-Physical_Facilities' : 'Pedestrian Crossing Physical Facilities', 
     'Light_Conditions' : 'Light Conditions', 
     'Weather_Conditions' : 'Weather Conditions', 
     'Road_Surface_Conditions' : 'Road Surface Conditions', 
     'Special_Conditions_at_Site' : 'Special Conditions at Site', 
     'Carriageway_Hazards' : 'Carriageway Hazards', 
     'Urban_or_Rural_Area' : 'Urban or Rural Area', 
     'Did_Police_Officer_Attend_Scene_of_Accident' : 'Police Attended Scene of Accident', 
     'LSOA_of_Accident_Location' : 'LSOA of Accident Location', 
     'Year' : 'Year', 
    })
              
```

```python
traffic_df['Date'] = pd.to_datetime(traffic_df['Date'], dayfirst=True)
traffic_df['Month'] = traffic_df['Date'].dt.month
traffic_df['Day'] = traffic_df['Date'].dt.day
```

```python
time_of_day = traffic_df.groupby(['Time']).sum()['Number_of_Casualties'].to_frame()
time_of_day = time_of_day.sort_index()
time_of_day.head()
```

```python
# convert 
def to_hour(time):
    try:
        hour = datetime.strptime(str(time), '%H:%M')
        return int(datetime.strftime(hour, '%H'))
    except Exception:
        return 0

traffic_df['Hour of Day'] = traffic_df['Time'].apply(to_hour)
```

```python
month_of_year = traffic_df.groupby(['Month']).sum()['Number_of_Casualties'].to_frame()
day_of_week = traffic_df.groupby(['Day_of_Week']).sum()['Number_of_Casualties'].to_frame()
hour_of_day = traffic_df.groupby(['Hour of Day']).sum()['Number_of_Casualties'].to_frame().reset_index(drop=True)
week_and_hour = traffic_df.groupby(['Day_of_Week','Hour of Day']).sum()['Number_of_Casualties'].to_frame().reset_index()
```

```python
week_and_hour.head()
```

```python
week_and_hour_pivot = week_and_hour.pivot_table(values=['Number_of_Casualties'], 
                                          index=['Hour of Day'], columns=['Day_of_Week'], aggfunc=np.sum)
week_and_hour_pivot['Number_of_Casualties'].head()
```

```python
plt.figure(figsize=(12,6))
plt.title("Time of Day and Traffic Accidents")
plt.plot(week_and_hour_pivot['Number_of_Casualties'][2], label='Monday')
plt.plot(week_and_hour_pivot['Number_of_Casualties'][3], label='Tuesday')
plt.plot(week_and_hour_pivot['Number_of_Casualties'][4], label='Wednesday')
plt.plot(week_and_hour_pivot['Number_of_Casualties'][5], label='Thursday')
plt.plot(week_and_hour_pivot['Number_of_Casualties'][6], label='Friday')
#plt.plot(week_and_hour_pivot['Number_of_Casualties'][1], label='Sunday')
#plt.plot(week_and_hour_pivot['Number_of_Casualties'][7], label='Saturday')
plt.legend()
```

```python
# convert 
traffic_df['Urban or Rural Area'] = [str('Urban') if value==1 else str('Rural') if value==2 else str('Neither') 
                                     for value in traffic_df['Urban_or_Rural_Area']]

area_road_type = traffic_df.groupby(['Urban or Rural Area','Road_Type']).sum()['Number_of_Casualties'].to_frame().reset_index()
```

```python
urban_road = area_road_type.loc[area_road_type['Urban or Rural Area']=='Urban',:]
#urban_road = urban_road.drop('Urban or Rural Area', axis=1).set_index('Road_Type')
rural_road = area_road_type.loc[area_road_type['Urban or Rural Area']=='Rural',:]
#rural_road = rural_road.drop('Urban or Rural Area', axis=1).set_index('Road_Type')
```

```python
rural_road
```

```python
urban_road.plot(kind='barh', label='Urban Casualties Count')
rural_road.plot(kind='barh', label='Rural Casualties Count')
```

```python
plt.pie(urban_road['Number_of_Casualties'], labels=urban_road['Road_Type'], 
        autopct='%.0f%%')
plt.title('Urban Traffic Casualties by Road Type')
plt.show()
```

```python
plt.pie(rural_road['Number_of_Casualties'], labels=rural_road['Road_Type'], 
        autopct='%.0f%%')
plt.title('Rural Traffic Casualties by Road Type')
plt.show()
```
